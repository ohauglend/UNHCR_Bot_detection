```python
## Function for going from.gz to df 

def gz_to_df(path, bot, source_label, compression='gzip', header=0, engine='python', sep='{"created_at"', quotechar='"'):

  """
  Datasets following this schema:
    botwiki-2019
    verified-2019
    cresci-stock-2018

  The function is used to go from zipped (.gz) files to data frame
    It is taylored to data from the botometer reposotory https://botometer.osome.iu.edu/bot-repository/datasets.html

  The function takes a file path, 
    bot 1/0 - where 1 is a bot to annotate wheter the content is a bot, 
    source_label a string that shows which sub dataset the data is from

  The function returns a df of the data source

  """

  #Reading in raw data
  df = pd.read_csv(path, compression=compression, header=header, engine=engine, sep=sep, quotechar=quotechar) 

  #Declaring variables
  created_at_1 = []
  user_id = []
  user_name =[]
  user_screen_name = []
  user_location = []
  user_description = []
  user_url = []
  protected = []
  followers_count = []
  friends_count = []
  listed_count = []
  created_at_2 = []
  favourites_count = []
  utc_offset = []
  time_zone = []
  geo_enabled = []
  verified = []
  statuses_count = []
  lang = []
  contributors_enabled = []
  is_translator = []
  is_translation_enabled = []
  profile_background_color = [] 
  profile_background_image_url = []
  profile_background_image_url_https = []
  profile_background_tile = []
  profile_image_url = []
  profile_image_url_https = []
  profile_link_color = []
  profile_sidebar_border_color = []
  profile_sidebar_fill_color = []
  profile_text_color = []
  profile_use_background_image = []
  following = []
  follow_request_sent = []
  notifications = []
  translator_type = []

  #Count is used to skip first entry
  count = 0

  #Looping through accounts and populating variables
  for acc in df:
    count +=1
    if count == 1:
      continue
    
    #Since '{"created_at_1"' is used to split the accounts we add this to the front of the acc entry
    acc = '{"created_at_1"' + acc

    #Populating variables on string entry - all variable population follows the same schema:
      #Forward split on varaible name and end split on next variable name - schema constructed from manual inspection
    created_at_1.append("".join(re.split(", \"user","".join(re.split("\"created_at_1\": ", acc)[1:]))[0:1]))
    user_id.append("".join(re.split(", \"id_str","".join(re.split("{\"id\": ",acc)[1:]))[0:1]))
    user_name.append("".join(re.split(", \"screen_name\": ","".join(re.split("\"name\": ",acc)[1:]))[0:1]))
    user_screen_name.append("".join(re.split(", \"location\": ","".join(re.split(", \"screen_name\": ",acc)[1:]))[0:1]))
    user_location.append("".join(re.split(", \"description\": ","".join(re.split(", \"location\": ",acc)[1:]))[0:1]))
    user_description.append("".join(re.split(", \"url\": ","".join(re.split(", \"description\": ",acc)[1:]))[0:1]))
    user_url.append("".join(re.split(", \"entities\":","".join(re.split(", \"url\": ",acc)[1:]))[0:1]))
    protected.append("".join(re.split(", \"followers_count","".join(re.split("\"protected\": ", acc)[1:]))[0:1]))
    followers_count.append("".join(re.split(", \"friends_count","".join(re.split("\"followers_count\": ", acc)[1:]))[0:1]))
    friends_count.append("".join(re.split(", \"listed_count","".join(re.split("\"friends_count\": ", acc)[1:]))[0:1]))
    listed_count.append("".join(re.split(", \"created_at","".join(re.split("\"listed_count\": ", acc)[1:]))[0:1]))
    created_at_2.append("".join(re.split(", \"favourites_count","".join(re.split("\"created_at\": ", acc)[1:]))[0:1]))
    favourites_count.append("".join(re.split(", \"utc_offset","".join(re.split("\"favourites_count\": ", acc)[1:]))[0:1]))
    utc_offset.append("".join(re.split(", \"time_zone","".join(re.split("\"utc_offset\": ", acc)[1:]))[0:1]))
    time_zone.append("".join(re.split(", \"geo_enabled","".join(re.split("\"time_zone\": ", acc)[1:]))[0:1]))
    geo_enabled.append("".join(re.split(", \"verified","".join(re.split("\"geo_enabled\": ", acc)[1:]))[0:1]))
    verified.append("".join(re.split(", \"statuses_count","".join(re.split("\"verified\": ", acc)[1:]))[0:1]))
    statuses_count.append("".join(re.split(", \"lang","".join(re.split("\"statuses_count\": ", acc)[1:]))[0:1]))
    lang.append("".join(re.split(", \"contributors_enabled","".join(re.split("\"lang\": ", acc)[1:]))[0:1]))
    contributors_enabled.append("".join(re.split(", \"is_translator","".join(re.split("\"contributors_enabled\": ", acc)[1:]))[0:1]))
    is_translator.append("".join(re.split(", \"is_translation_enabled","".join(re.split("\"is_translator\": ", acc)[1:]))[0:1]))
    is_translation_enabled.append("".join(re.split(", \"profile_background_color","".join(re.split("\"is_translation_enabled\": ", acc)[1:]))[0:1]))
    profile_background_color.append("".join(re.split(", \"profile_background_image_url","".join(re.split("\"profile_background_color\": ", acc)[1:]))[0:1]))
    profile_background_image_url.append("".join(re.split(", \"profile_background_image_url_https","".join(re.split("\"profile_background_image_url\": ", acc)[1:]))[0:1]))
    profile_background_image_url_https.append("".join(re.split(", \"profile_background_tile","".join(re.split("\"profile_background_image_url_https\": ", acc)[1:]))[0:1]))
    profile_background_tile.append("".join(re.split(", \"profile_image_url","".join(re.split("\"profile_background_tile\": ", acc)[1:]))[0:1]))
    profile_image_url.append("".join(re.split(", \"profile_image_url_https","".join(re.split("\"profile_image_url\": ", acc)[1:]))[0:1]))
    profile_image_url_https.append("".join(re.split(", \"profile_link_color","".join(re.split("\"profile_image_url_https\": ", acc)[1:]))[0:1]))
    profile_link_color.append("".join(re.split(", \"profile_sidebar_border_color","".join(re.split("\"profile_link_color\": ", acc)[1:]))[0:1]))
    profile_sidebar_border_color.append("".join(re.split(", \"profile_sidebar_fill_color","".join(re.split("\"profile_sidebar_border_color\": ", acc)[1:]))[0:1]))
    profile_sidebar_fill_color.append("".join(re.split(", \"profile_text_color","".join(re.split("\"profile_sidebar_fill_color\": ", acc)[1:]))[0:1]))
    profile_text_color.append("".join(re.split(", \"profile_use_background_image","".join(re.split("\"profile_text_color\": ", acc)[1:]))[0:1]))
    profile_use_background_image.append("".join(re.split(", \"following","".join(re.split("\"profile_use_background_image\": ", acc)[1:]))[0:1]))
    following.append("".join(re.split(", \"follow_request_sent","".join(re.split("\"following\": ", acc)[1:]))[0:1]))
    follow_request_sent.append("".join(re.split(", \"notifications","".join(re.split("\"follow_request_sent\": ", acc)[1:]))[0:1]))
    notifications.append("".join(re.split(", \"translator_type","".join(re.split("\"notifications\": ", acc)[1:]))[0:1]))
    translator_type.append("".join(re.split("\"translator_type\": ", acc)[1:])[:-4])
    
  #creating df
  df2 = pd.DataFrame(
      {'created_at_1': created_at_1,
      'user_id': user_id,
      'user_name': user_name,
      'user_screen_name': user_screen_name,
      'user_location': user_location,
      'user_description': user_description, 
      'user_url': user_url, 
      'protected': protected,
      'followers_count': followers_count,
      'friends_count': friends_count,
      'listed_count': listed_count,
      'created_at_2': created_at_2,
      'favourites_count': favourites_count,
      'utc_offset': utc_offset,
      'time_zone': time_zone,
      'geo_enabled': geo_enabled,
      'verified': verified,
      'statuses_count': statuses_count,
      'lang': lang,
      'contributors_enabled': contributors_enabled,
      'is_translator': is_translator,
      'is_translation_enabled': is_translation_enabled,
      'profile_background_color': profile_background_color,
      'profile_background_image_url': profile_background_image_url,
      'profile_background_image_url_https': profile_background_image_url_https,
      'profile_background_tile': profile_background_tile,
      'profile_image_url': profile_image_url,
      'profile_image_url_https': profile_image_url_https,
      'profile_link_color': profile_link_color,
      'profile_sidebar_border_color': profile_sidebar_border_color,
      'profile_sidebar_fill_color': profile_sidebar_fill_color,
      'profile_text_color': profile_text_color,
      'profile_use_background_image': profile_use_background_image,
      'following': following,
      'follow_request_sent': follow_request_sent,
      'notifications': notifications,
      'translator_type': translator_type
      })

  #Labeling according to input
  df2["Bot"] = [bot for _ in range(len(df2))]
  df2["Source/label"] = [source_label for _ in range(len(df2))]

  return df2
```
